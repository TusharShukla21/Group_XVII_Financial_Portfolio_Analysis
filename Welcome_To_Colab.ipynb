```
import pandas as pd
import matplotlib.pyplot as plt

file_path = "/content/yahoo_data.xlsx"

data = pd.read_excel(file_path)

print(data.head())

if 'Adj Close**' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])

    data['Daily Return'] = data['Adj Close**'].pct_change()

    data.dropna(inplace=True)

    data['Year'] = data['Date'].dt.year
    grouped_data = data.groupby('Year')['Daily Return']


    risk_per_year = grouped_data.std()
    return_per_year = grouped_data.mean()


    summary = pd.DataFrame({'Risk': risk_per_year, 'Return': return_per_year})
    print(summary)


    summary.plot(kind='bar', figsize=(10, 6), color=['b', 'g'], width=0.8)


    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title('Risk and Return by Year')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(['Risk (Standard Deviation)', 'Return (Mean)'])


    plt.show()
else:
    print("Error: The 'Adj Close**' column was not found in the Excel file.")
```
